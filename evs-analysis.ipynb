```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("/kaggle/input/evs-one-electric-vehicle-dataset/ElectricCarData_Clean.csv") 
```

```python
data.head()
```

```python
data = data.rename(columns={'Model': 'Vehicle_type', 'TopSpeed_KmH': 'Average_Speed','PriceEuro':'Average_price','Brand':'Company'})
data.drop(['PowerTrain','PlugType'], axis=1)

```

```python
data.info()
```

```python
data = pd.DataFrame(data) 

data = data[~data['FastCharge_KmH'].isin(['-'])] 
# removing the records where the fast charging is unavailable

data['FastCharge_KmH'] = data['FastCharge_KmH'].apply(pd.to_numeric) 
# changing the fast charging values from object to numeric so that we would be able to visualize it

data.head(100) 
```

```python
import plotly.express as px 
fig = px.box(data,x='Average_price',y ='Company' , color='Company', points='all', hover_data=['Vehicle_type'] ,
           height=1000, width= 900, orientation='h', title='Price and Brand' , template='ggplot2' ,
            labels=dict(PriceEuro='Price (€)'))
fig.show()
```

We can see from this graph that Porsche have the most expensive vehicles in the market with the average price of 148.301k Euro, but the Tesla Roadster is the most expensive with the price 215k Euro and the SEAT Mii Electric is the most affordable with the price 20.129k Euro.

```python
fig = px.scatter(data,x = 'Average_Speed',y = 'AccelSec', color = 'Company', hover_data=['Vehicle_type'] ,
           opacity=0.8 , title='Acceleration and Top Speed' , height=650 , width=900 , template='ggplot2' ,
                labels=dict(AccelSec='Acceleration (Sec)', TopSpeed_KmH='Top Speed (Km/H)'))

fig.update_traces(marker={'size': 8}) # adjusting the markers size
fig.show()
```

From here it cane be concluded that the Tesla Roadster can go from 0 to 100 Km/H in 2.1 Seconds and its top speed is 410 Km/H, which make it the best Vehicle in terms of top speed and acceleration, with most of the vehicles lying between 140-260 Km/H top speed and 2-10 Seconds acceleration.

```python
fig = px.scatter(data, x = 'FastCharge_KmH', y = 'Range_Km', color = 'Company', hover_data=['Vehicle_type'] ,
           opacity=0.8 , title='Range and Fast Charging' , height=650 , width=900 , template='ggplot2' ,
                labels=dict(Range_Km='Range (Km)', FastCharge_KmH='Fast Charging (Km/H)'))

fig.update_traces(marker={'size': 8}) # adjusting the markers size
fig.show()
```

We can see here that in general, Tesla and Porscha vehicle have the best charging rates ranging between 700-970 Km/H, with Tesla topping in the range with some of their vehicles that can go over 500 Km with a full battery.

```python
fig = px.histogram(data, x='Range_Km', y='Average_price', color='Company', barmode='group', histfunc='avg', 
                   height=400 , width=1100 , nbins=10 , template='ggplot2' , title='Average Price and Range' ,
                    labels=dict(PriceEuro='Price (€)', Range_Km='Range (Km)'))

fig.show()
```

This histogram shows the average price of each compny's vehicle in a certain range. We can see on here that most of the vehicles ranging 200-500 Km with their price going between 25-150k Euro depending on the company. Some vehicles can go between 100-200 Km with a single charge and priced between 20k-35k Euro which make them slightly cheeper than the others.
